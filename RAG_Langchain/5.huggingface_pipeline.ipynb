### Huggingface transformers libraray

Hugging Face’s pipeline is one of the easiest entry points into using their Transformers library. It abstracts away most of the complexity and lets you quickly run state-of-the-art models for tasks like text classification, translation, summarization, question answering, and more


```python
from huggingface_hub import login
from transformers import pipeline
```

```python
import os
hf_token = os.getenv('HUGGINGFACEHUB_API_TOKEN')
```

```python
# NOTE: When you have not specified the model name, the default model is used. And the model will be downloaded on your machine 

classifier = pipeline("sentiment-analysis", device="cuda")
result = classifier("I'm super excited to be on the way to LLM mastery!")
print(result)
```

```python
# Named Entity Recognition

ner = pipeline("ner", grouped_entities=True, device="cuda")
result = ner("Hugging Face is a technology company based in New York and Paris.")
print(result)
```

```python
# Question Answering with Context

question_answerer = pipeline("question-answering", device="cuda")
result = question_answerer(question="What is the name of the company?", context="Hugging Face is a technology company based in New York and Paris.")
print(result)
```

```python
# Text Generation

generator = pipeline("text-generation", device="cuda")
result = generator("Once upon a time in a land far, far away", max_length=50, num_return_sequences=1)
print(result[0]['generated_text'])
```
